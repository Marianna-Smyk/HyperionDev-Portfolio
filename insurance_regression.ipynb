```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

```python
# Plot a scatter plot with age on the x-axis and charges on the y-axis
x = insurance_data.loc[:, 'age'].values
y = insurance_data.loc[:, 'charges'].values

plt.scatter(x, y)
plt.title('Scatter Plot of Age vs Charges')
plt.xlabel('Age (years)')
plt.ylabel('Charges ($)')
plt.show()
```

```python
# Prepare the data for linear regression
X = insurance_data['age'].values.reshape(-1, 1)  # create a 2D array
y = insurance_data['charges'].values
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Fit the model to the training data
insurance_costs_model = LinearRegression()
insurance_costs_model.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = insurance_costs_model.predict(X_test)
```

```python
# Plot another scatter plot with the best-fit line
plt.scatter(X_test, y_test, label='Actual data')
plt.plot(X_test, y_pred, color = 'r', label='Best-fit line')
plt.title('Scatter Plot of Age vs Charges with the Best-fit Line')
plt.xlabel('Age (years)')
plt.ylabel('Charges ($)')
plt.legend()
plt.show()
```

```python
# Predict an unknown value (example)
X_test = 15
y_pred = insurance_costs_model.predict([[X_test]])
print(f"Insurance cost for {X_test}-year-olds should be: {y_pred}$")
```
